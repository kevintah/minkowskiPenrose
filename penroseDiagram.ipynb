```python
from sympy.deprecated.class_registry import C
from sympy.plotting import plot
from sympy.core.symbol import symbols
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# symbols#du*dv = 0
#so constants 
#u = C1
#v = C2

ticks = (-0.5*np.pi, 0.5*np.pi, 0.5)
u  = np.arange(-0.5*np.pi, 0.5*np.pi, 0.5)
v  = np.arange(-0.5*np.pi, 0.5*np.pi, 0.5)


t = np.arange(-0.5*np.pi, 0.5*np.pi, 0.5)
r = np.arange(-0.5*np.pi, 0.5*np.pi, 0.5)

T = np.arange(-1000,1000, 0.01)
R = np.arange( 0,2000, 0.01)

for a in range(0, u.size):
  plt.xlabel('r')
  plt.ylabel('t')
  plt.plot(r,u[a]+r,'red',r,v[a]-r,'blue')
plt.title('Minkowski')
plt.show()


for a in range(0, r.size):
  plt.xlabel('R')
  plt.ylabel('T')
  plt.plot(
           (np.arctan(T + a*0.5) - np.arctan(T-a*0.5)), (np.arctan(T - a*0.5) + np.arctan(T+a*0.5)),'red',
           (np.arctan(T + a*0.1) - np.arctan(T-a*0.1)), (np.arctan(T - a*0.1) + np.arctan(T+a*0.1)),'red',
           (np.arctan(a*0.5 + R) - np.arctan(a*0.5-R)), (np.arctan(a*0.5 - R) + np.arctan(a*0.5+R)),'blue',
           (np.arctan(-a*0.5 + R) - np.arctan(-a*0.5-R)), (np.arctan(-a*0.5 - R) + np.arctan(-a*0.5+R)),'blue'
           )
plt.title('Minkowski Penrose diagram  -- static - compact restricted')       
plt.show()
```
